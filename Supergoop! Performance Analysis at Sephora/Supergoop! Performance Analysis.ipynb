**Conclusion: Columns with Identified Typographical Issues and Inconsistencies**


1. size: Inconsistent formatting for the same size values due to variations in spacing and separators.

2. variation_value: Mixed content types and inconsistent formatting, including differences in capitalization and spacing.

3. variation_desc: Minor typographical errors and inconsistent capitalization across descriptive values.





```python
# Fix typographical issues & inconsistencies in categorical columns

# 1. SIZE — standardize spacing & separators
df_product['size'] = (
    df_product['size']
    .str.strip()
    .str.replace(r'\s*/\s*', ' / ', regex=True)   # standardize slash spacing
    .str.replace(r'\s+', ' ', regex=True)         # remove extra spaces
    .str.replace(r'^\.','0.', regex=True)         # fix leading decimal (e.g. .11 -> 0.11)
)

# 2. variation_value
df_product['variation_value'] = (
    df_product['variation_value']
    .astype(str)
    .str.strip()
    .str.replace(r'\s*/\s*', ' / ', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
    .replace(['Nan', 'nan', 'None', ''], 'Unknown')
)

# 3. VARIATION_DESC — fix casing & known typos
df_product['variation_desc'] = (
    df_product['variation_desc']
    .str.strip()
    .str.lower()
    .str.replace('fuschia', 'fuchsia', regex=False)  # fix known typo
    .str.title()
)
```

**Conclusion**

Typographical and categorical inconsistencies were identified in the **review_title, eye_color, and skin_tone columns.** These issues mainly involve variations in capitalization, punctuation, and naming conventions (e.g., “Love it” vs “Love it!”, “Grey” vs “gray”, and mixed-case labels such as lightMedium or notSureST)

```python
# Fix typographical issues & inconsistencies in categorical columns
# ============================================

# 1) review_title: normalize punctuation & capitalization
df_review['review_title'] = (
    df_review['review_title']
    .str.lower()
    .str.replace(r'[!]+', '', regex=True)
    .str.strip()
)

# 2) eye_color: fix capitalization inconsistency
df_review['eye_color'] = (
    df_review['eye_color']
    .str.lower()
    .replace({'grey': 'gray'})
)

# 3) skin_tone: standardize naming format (camelCase & unclear labels)
df_review['skin_tone'] = (
    df_review['skin_tone']
    .str.lower()
    .replace({
        'lightmedium': 'light medium',
        'fairlight': 'fair light',
        'mediumtan': 'medium tan',
        'notsurest': 'not sure'
    })
)

```

### Checking Outlier

```python
# Kolom numerik magnitude (unbounded / skewed)
num_cols = [
    'price_usd_x',
    'price_usd_y',
    'sale_price_usd',
    'value_price_usd',
    'loves_count',
    'reviews',
    'rating_x',
    'helpfulness',
    'total_feedback_count',
    'total_pos_feedback_count',
    'total_neg_feedback_count',
    'child_min_price',
    'child_max_price'
]

print("Outlier Summary (IQR Method):\n")

for col in num_cols:
    Q1 = df_merged[col].quantile(0.25)
    Q3 = df_merged[col].quantile(0.75)
    IQR = Q3 - Q1
    LB  = Q1 - 1.5 * IQR
    UB  = Q3 + 1.5 * IQR

    outlier_count = ((df_merged[col] < LB) | (df_merged[col] > UB)).sum()
    print(f"{col}: {outlier_count} outliers")
```

**Insight:**
 The IQR-based outlier analysis identified a small number of outliers in pricing variables (price_usd_x, price_usd_y, value_price_usd), product popularity (loves_count, reviews), and review-related metrics (rating_x, helpfulness, total_pos_feedback_count, total_neg_feedback_count). These outliers represent valid business behavior rather than data quality issues, so all observations were retained for further analysis

```python
# Removing/Excluding the outlier and assign it on new variable "df_clean" by using negate (~)
df_clean = df_merged.copy()

for col in num_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    LB = Q1 - 1.5 * IQR
    UB = Q3 + 1.5 * IQR

    df_clean = df_clean[
        ~((df_clean[col] < LB) | (df_clean[col] > UB))
    ]

print("Before:", df_merged.shape)
print("After :", df_clean.shape)
```

### Preparing Data

```python
df_clean['positive_ratio'] = (
    df_clean['total_pos_feedback_count'] /
    (df_clean['total_feedback_count'] + 1e-6)
)

df_clean['negative_ratio'] = (
    df_clean['total_neg_feedback_count'] /
    (df_clean['total_feedback_count'] + 1e-6)
)

df_prepared = df_clean[[
    'price_usd_y',        # pricing
    'rating_x',           # rating
    'reviews',            # exposure / popularity
    'positive_ratio',     # sentiment
    'negative_ratio'      # sentiment
]]

df_prepared
```

**Insight:** The data shows minimal variation in pricing, ratings, and review counts, with most differentiation coming from extreme positive or negative sentiment ratios. As a result, clustering will primarily capture sentiment-based patterns rather than product-level differences.

### Robust Scaler

```python
from sklearn.preprocessing import RobustScaler

columns = [
    'price_usd_y',
    'rating_x',
    'reviews',
    'positive_ratio',
    'negative_ratio'
]

rs_scaler = RobustScaler()

df_scaled = df_prepared.copy()
df_scaled[columns] = rs_scaler.fit_transform(df_scaled[columns])

df_scaled[columns].describe()
```

**Insight:** The RobustScaler results show that only negative_ratio exhibits variation, while the other features remain constant. As a result, the clustering analysis will primarily reflect differences in negative sentiment rather than product characteristics.

### Elbow Method

>**Elbow Method helps identify how many meaningful customer segments should be created so Sephora can design targeted strategies to improve ratings and online sales.**

```python
from sklearn.cluster import KMeans

inertia = []

for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(6,4))
plt.plot(range(1, 6), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.xticks(range(1, 6))
plt.show()
```

**Insight:** The elbow method indicates an optimal number of clusters at k = 2, suggesting a clear separation driven by user sentiment. This result confirms that negative feedback is the primary factor differentiating products, rather than pricing or popularity.

### Descriptive Statistic

```python
df_prepared = df_prepared.copy()
df_prepared['cluster'] = KMeans(n_clusters=2, random_state=42).fit_predict(df_scaled)

cluster_stats = (
    df_prepared
    .groupby('cluster')
    .agg(['mean', 'median', 'min', 'max'])
)

cluster_stats.columns = [f"{c}_{s}" for c, s in cluster_stats.columns]
cluster_stats = cluster_stats.reset_index().round(3)

cluster_stats
```
